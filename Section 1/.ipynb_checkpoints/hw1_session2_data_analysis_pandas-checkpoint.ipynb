**<font color='red'>Вопрос 6 (2 балла).</font> Постройте новый признак – BMI ([Body Mass Index](https://en.wikipedia.org/wiki/Body_mass_index)). Для этого надо вес в килограммах поделить на квадрат роста в метрах. Нормальными считаются значения BMI от 18.5 до 25. Выберите верные утверждения.**

**<font color='red'>Утверждения:</font>**
- Медианный BMI по выборке лежит в пределах нормы
- У женщин в среднем BMI выше, чем у мужчин
- У здоровых в среднем BMI выше, чем у больных
- В сегменте здоровых и непьющих мужчин BMI ближе к норме, чем в сегменте здоровых и непьющих женщин

```python
df['BMI'] = df['weight']/((df['height']/100) * (df['height']/100))
```

```python
df['BMI'].median()
#Медианный BMI по выборке лежит в пределах нормы - НЕТ
```

```python
df[df['gender'] == 1]['BMI'].mean(), \
df[df['gender'] == 2]['BMI'].mean()
#У женщин в среднем BMI выше, чем у мужчин - ДА
```

```python
df[df['cardio'] == 0]['BMI'].mean(), \
df[df['cardio'] == 1]['BMI'].mean()
#У здоровых в среднем BMI выше, чем у больных - НЕТ
```

```python
df[(df['cardio'] == 0) & (df['gender'] == 2) & (df['alco'] == 0)]['BMI'].mean(), \
df[(df['cardio'] == 0) & (df['gender'] == 1) & (df['alco'] == 0)]['BMI'].mean()
#В сегменте здоровых и непьющих мужчин BMI ближе к норме, чем в сегменте здоровых и непьющих женщин - Да
```
